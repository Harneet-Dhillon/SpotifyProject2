```python
import pandas as pd
```

```python
#importing all the json files of Spotify data

df1 = pd.read_json('StreamingHistory_music_2.json')
df1
```

```python
df2 = pd.read_json('StreamingHistory_music_3.json')
df2
```

```python
breakup = pd.concat([df1,df2])
breakup
```

```python
#converting the json file into csv

breakup.to_csv("breakup.csv",index=False)
thebreakup = pd.read_csv("breakup.csv")
thebreakup
```

```python
#finding my top 20 most played artists during my breakup era 
artist_count = thebreakup["artistName"].value_counts()
thebreakup["artistCount"] = thebreakup["artistName"].map(artist_count)
Fav_artists = thebreakup[["artistName", "artistCount"]]
Fav_artists = Fav_artists.drop_duplicates()
Fav_artists_sorted = Fav_artists.sort_values(by="artistCount", ascending=False)
Fav_artists_sorted  = Fav_artists_sorted.reset_index(drop=True)
Fav_artists_sorted = Fav_artists_sorted.head(20)
Fav_artists_sorted
```

```python
#finding average number of songs played per day during my breakup era 
thebreakup[["date", "time"]] = thebreakup['endTime'].str.split(' ', expand=True)
date_count = thebreakup["date"].value_counts()

songs_everyday = pd.DataFrame(thebreakup["date"])
songs_everyday["dateCount"] = thebreakup["date"].map(date_count)
songs_everyday = songs_everyday.drop_duplicates()
songs_everyday = songs_everyday.drop(0)
songs_everyday  = songs_everyday.reset_index(drop=True)
songs_everyday
```

```python
avg = songs_everyday["dateCount"].mean()
print(avg)
#avg = 202 songs per day
```

```python
#finding top 15 most played songs during my breakup era 
#create new column = unique od artistname + trackname and count on that 
thebreakup["uniqueID"] = thebreakup["trackName"]+ " by " + thebreakup["artistName"]

song_count = thebreakup["uniqueID"].value_counts()
thebreakup["songCount"] = thebreakup["uniqueID"].map(song_count)
songs_table = thebreakup[["uniqueID", "songCount"]]
songs_table = songs_table.drop_duplicates()
songs_table = songs_table.sort_values(by="songCount", ascending=False)
songs_table  = songs_table.reset_index(drop=True)
songs_table= songs_table.head(15)
songs_table
```

```python
#need to export all these as csv files to put into tableau 

#table for top 20 artists
Fav_artists_sorted.to_csv('Top20Artistsbreakup1.csv')

#total songs listened to everyday
songs_everyday.to_csv('TotalSongsEverydaybreakup.csv')

#top 15 songs most listened to 
songs_table.to_csv('Top15Songsbreakup1.csv')

print('done')
```
