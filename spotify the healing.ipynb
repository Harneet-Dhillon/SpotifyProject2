```python
import pandas as pd
```

```python
#importing all the json files of Spotify data

df3 = pd.read_json('StreamingHistory_music_4.json')
df3
```

```python
df4 = pd.read_json('StreamingHistory_music_5.json')
df4
```

```python
df5 = pd.read_json('StreamingHistory_music_6.json')
df5
```

```python
healing = pd.concat([df3,df4,df5])
healing
```

```python
#converting the json file into csv


healing.to_csv("healing.csv",index=False)
thehealing = pd.read_csv("healing.csv")
thehealing
```

```python
#finding my top 20 most played artists during my healing era 

artist_count = thehealing["artistName"].value_counts()
thehealing["artistCount"] = thehealing["artistName"].map(artist_count)
Fav_artists = thehealing[["artistName", "artistCount"]]
Fav_artists = Fav_artists.drop_duplicates()
Fav_artists_sorted = Fav_artists.sort_values(by="artistCount", ascending=False)
Fav_artists_sorted  = Fav_artists_sorted.reset_index(drop=True)
Fav_artists_sorted = Fav_artists_sorted.head(20)
Fav_artists_sorted
```

```python
#finding average number of songs played per day during my healing era 

thehealing[["date", "time"]] = thehealing['endTime'].str.split(' ', expand=True)
date_count = thehealing["date"].value_counts()

songs_everyday = pd.DataFrame(thehealing["date"])
songs_everyday["dateCount"] = thehealing["date"].map(date_count)
songs_everyday = songs_everyday.drop_duplicates()
songs_everyday = songs_everyday.drop(0)
songs_everyday  = songs_everyday.reset_index(drop=True)
songs_everyday
```

```python
avg = songs_everyday["dateCount"].mean()
print(avg)
#avg = 171 songs per day
```

```python
#finding top 15 most played songs during my healing era 
#create new column = unique od artistname + trackname and count on that 
thehealing["uniqueID"] = thehealing["trackName"]+ " by " + thehealing["artistName"]

song_count = thehealing["uniqueID"].value_counts()
thehealing["songCount"] = thehealing["uniqueID"].map(song_count)
songs_table = thehealing[["uniqueID", "songCount"]]
songs_table = songs_table.drop_duplicates()
songs_table = songs_table.sort_values(by="songCount", ascending=False)
songs_table = songs_table[songs_table["uniqueID"].where(songs_table["uniqueID"] != "Unknown Track by Unknown Artist").notna()]

songs_table  = songs_table.reset_index(drop=True)
songs_table = songs_table.head(15)
songs_table
```

```python
#need to export all these as csv files to put into tableau 

#table for top 20 artists
Fav_artists_sorted.to_csv('Top20Artistshealing1.csv')

#total songs listened to everyday
songs_everyday.to_csv('TotalSongsEverydayhealing.csv')

#top 15 songs most listened to 
songs_table.to_csv('Top15Songshealing1.csv')

print('done')
```
